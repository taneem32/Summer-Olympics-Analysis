### 1. In how many cities Summer Olympics is held so far?

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
# Code without using any inbuilt function
data = []
for sport in df['Sport'].unique():
    df1 = df[df['Sport'] == sport]
    data.append([sport,len(df1[df1['Medal'] == 'Gold'])])
    
pd.DataFrame(data,columns=['Sport','Freq_Of_Gold']).sort_values(by = 'Freq_Of_Gold',ascending = False).head().plot(x='Sport',y='Freq_Of_Gold',kind = 'bar',figsize = (8,5))
  
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
#BELOW IS THE CODE WITHOUT USING INBUILT FUNCTIONS
data = []
for player in df['Athlete'].unique():
    df1 = df[df['Athlete'] == player]
    data.append([player,len(df1[df1['Medal'] == 'Gold'])])
    
pd.DataFrame(data,columns=['Player','Gold_count']).sort_values(by='Gold_count',ascending = False).head().plot(x='Player',y='Gold_count',kind = 'bar',figsize=(8,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = []
df1 = df[df['Country'] == 'IND']
df2 = df1[df1['Medal'] == 'Gold']
for year in df2['Year'].unique():
    data.append([year,df2['Medal']])
       
df2.sort_values(by='Year')
# 1928 -- India won first gold medal
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []
for sport in df['Sport'].unique():
    df1 = df[df['Sport'] == sport]
    df1 = df1[df1['Gender'] == 'Women']
    data.append([sport,len(df1['Medal'] == 'Gold')])
pd.DataFrame(data,columns=['Sport','Female_Gold']).sort_values(by='Female_Gold',ascending = False).head().plot(x='Sport',y='Female_Gold',kind = 'bar',figsize=(8,5))
```
